## Correlations across the dataset

There are little correlations throughout the dataset. Nevertheless, let us take a look at a few variables that are somewhat correlated to the `Exited` column, which indicates if a customer has remained with bank or left it, namely:
- `Gender`.
- `Geography`.
- `IsActiveMember`.
- `Age`.
- `Balance`.

We will also take a look at `Tenure`, which indicates the number of years for which a customer has been with the bank. This will facilitate disentangling the age from the time spent as a customer at the bank (although we already see in the above figure that `Age` is more correlated to `Exited` than `Tenure`). 

## Churn likelihood for relevant categorical variables

```python
def get_exit_likelihood(df: pd.DataFrame,
                        column_name: str,
                        *,
                        sort_final: bool = True) -> pd.DataFrame:
    """
    Generate a table grouped by column_name with:
        - the total customer count ('total_count').
        - the count of customers that have exited ('exit_count').
        - the likelihood for customers to have exited ('exit_likelihood'),
            i.e., exit_count/total_count.

    If sort_final is True, the final dataframe is sorted by descending exit likelihood.
    """

    # Compute total count
    df_count = df[column_name].value_counts().copy()
    df_count = df_count.reset_index(name='total_count')

    # Compute exit count
    df_exit = df[df['Exited'] == 1].groupby(column_name).size()
    df_exit = df_exit.reset_index(name='exit_count')

    # Merge tables
    df_final = pd.merge(df_count, df_exit, on=column_name, how='left')

    # Compute exit proportions
    df_final['exit_likelihood'] = 100*df_final['exit_count']/df_final['total_count']

    # Sort according to exit proportion
    if sort_final:
        df_final = df_final.sort_values('exit_likelihood', ascending=False)
        df_final = df_final.reset_index(drop=True)

    return df_final


df_gender = get_exit_likelihood(df, 'Gender')
df_geography = get_exit_likelihood(df, 'Geography')
df_member = get_exit_likelihood(df, 'IsActiveMember')
```

```python
# For plotting
membership = ['Not Active Member' if x==0 else 'Active Member'
              for x in df_member['IsActiveMember'].to_list()]
df_member['active_member'] = membership

df_member
```

```python
# Plot
fig, ax = plt.subplots(3, 1)

sns.barplot(df_gender, x='exit_likelihood', y='Gender', ax=ax[0],
            hue='Gender', palette='dark:b_r')
# Add values to bars
for idx, val in enumerate(df_gender['exit_likelihood']):
        ax[0].text(val+0.5, idx, f'{val:.1f}%', va='center')
ax[0].set_xlim([0, 40])
ax[0].set_xticklabels([])
ax[0].set_xlabel('')
ax[0].set_ylabel('')

sns.barplot(df_geography, x='exit_likelihood', y='Geography', ax=ax[1],
            hue='Geography', palette='dark:b_r')
for idx, val in enumerate(df_geography['exit_likelihood']):
        ax[1].text(val+0.5, idx, f'{val:.1f}%', va='center') 
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_xticklabels([])
ax[1].set_xlabel('')
ax[1].set_ylabel('')

sns.barplot(df_member, x='exit_likelihood', y='active_member', ax=ax[2],
            hue='active_member', palette='dark:b_r')
for idx, val in enumerate(df_member['exit_likelihood']):
        ax[2].text(val+0.5, idx, f'{val:.1f}%', va='center') 
ax[2].set_xlim(ax[0].get_xlim())
ax[2].set_ylabel('')
ax[2].xaxis.set_major_formatter(PercentFormatter(decimals=0))
ax[2].set_xlabel('Churn Likelihood')

plt.suptitle('Churn Likelihood for Relevant Categorical Variables',
             fontsize=15)
plt.tight_layout(h_pad=1.5)
plt.savefig(os.path.join(output_directory, '1_2_churn_likelihoood_categorical.png'),
            bbox_inches='tight', dpi=600)
plt.show()
```

**Breakdown:**
- There is a larger proportion of female customers than male customers that have left the bank.
- There are twice as many German customers that have left the bank as there are Spanish or French ones.
- Customers that have left the bank were less likely to be active members.

## Distributions of customers for relevant continuous variables

```python
df_plot = df.copy()

df_plot['Exited'] = df_plot['Exited'].replace(0, 'Customers who Stayed')
df_plot['Exited'] = df_plot['Exited'].replace(1, 'Customers who Left')

fig, ax = plt.subplots(3, 1)

sns.boxplot(df_plot, x='Age', y='Exited',
            orient='h', ax=ax[0], color=[1, 0.5, 0])
ax[0].invert_yaxis()
ax[0].set_ylabel('')

sns.boxplot(df_plot, x='Tenure', y='Exited',
            orient='h', ax=ax[1], color=[1, 0.5, 0])
ax[1].invert_yaxis()
ax[1].set_ylabel('')
ax[1].set_xlabel('Years of Tenure')

sns.boxplot(df_plot, x='Balance', y='Exited',
            orient='h', ax=ax[2], color=[1, 0.5, 0])
ax[2].invert_yaxis()
ax[2].set_ylabel('')
ax[2].set_xlabel('Account Balance')
xticks_format = plt.FuncFormatter(lambda y, pos: f'{int(y/1000)}k')
plt.gca().xaxis.set_major_formatter(xticks_format)

plt.suptitle('Customer Distributions for Relevant Continuous Variables', fontsize=15)
plt.tight_layout(h_pad=1.8)
plt.savefig(os.path.join(output_directory, '1_3_customer_distributions_continuous.png'),
            bbox_inches='tight', dpi=600)
plt.show()
```

**Breakdown:**
- Customers who left the bank were typically older than customers who stayed.
- This age discrepancy does not seem to be related to the years of tenure, which are uncorrelated with leaving the bank.
    - This is also apparent in the correlation heatmap above.
- Customers who left the bank typically had larger balance in their accounts.
